```python
#create daily grid maps 
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
from pathlib import Path

grazer = [0,1,3,6,10,20,25,30,35]
for j in grazer:
    data = Path(f"data/model_runs/scenario_kp10/scenario/p2-g{j}/biomass_data.nc")
    varpar= data.parent/'daily_biomass_grid'
    varpar.mkdir(exist_ok=True)
 
    biomass_grids = xr.open_dataarray(data)
    #biomass_grids.coords['days'] = biomass_grids['time'] / 1440
    time= list(range(57))
    for i in time:
        outpath = varpar/f'day_{i:02d}.jpg'
        plt.imsave(outpath, biomass_grids[48*i], vmin=0, vmax=5,)
```

```python
#SEPARATE PLOTS p15
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr

data = "data/model_runs/scenario_kp10/scenario/p15-g0/biomass_data.nc"
 
biomass_grids = xr.open_dataarray(data)
#biomass_grids.coords['days'] = biomass_grids['time'] / 1440
tim= list(range(57))
for i in tim:
    plt.imshow(biomass_grids[48*i], vmin=-1, vmax=1, cmap='RdBu')
    plt.xlabel('grid dimension')
    plt.ylabel('grid dimension')
    plt.title(f'P=15, G=0 (day {i})')
    #plt.colorbar()
    plt.savefig(f"data/model_runs/scenario_kp10/scenario/grid/p15/p15-g0/grid_maps_p15-g0_day{i}.jpg" )
```
